```python
# Creating an aggregate of 15 min intervals
interval_logins = logins.resample('15T', on='login_time').count()
interval_logins
```

```python
interval_logins.login_time.mean()
```

```python
plt.figure(figsize=(16,8))

plt.plot(interval_logins, alpha=0.5)
plt.ylabel('Aggregate Count', fontsize=12)
plt.xlabel('Range of Time', fontsize=12)
plt.show()
```

From the above line plot, the following points can be understood:
    1. There seems to be a steady increase of logins over time. (upward trend)
    2. Weekly peak is observed.
    3. The mean logins for a certain period of time is concentrated at around 10.
  
    

```python
# Line Plot to understand the daily trend of logins

plt.figure(figsize=(16,8))
plt.plot(interval_logins.iloc[:100] , alpha=0.5)
plt.ylabel('Aggregate Count', fontsize=12)
plt.xlabel('Range of Time', fontsize=12)
plt.show()
```

```python
# The same plot visualized for 3 days to validate the above result

plt.figure(figsize=(16,8))
plt.plot(interval_logins.iloc[:300] , alpha=0.5)
plt.ylabel('Aggregate Count', fontsize=12)
plt.xlabel('Range of Time', fontsize=12)
plt.show()
```

From the daily trend plot visualized above, There are two peaks observed every day.
    1. Mid-night (between 12am and 03am)
    2. Noon (12pm)

There is an abrupt fall of logins just after the peak and they steadily rise till the next peak.

```python
# Checking to see which day of week the weekly peaks fall on

plt.figure(figsize=(16,8))

plt.plot(interval_logins.iloc[:1000] , alpha=0.5)
plt.ylabel('Aggregate Count', fontsize=12)
plt.xlabel('Range of Time', fontsize=12)
plt.show()
```

From the dates given, all the weekly peaks fall on a sunday

## User Retension

```python
from datetime import timedelta
```

```python
#User retension in the 6th month means that the user has to be active for more than 5 months

def user_retension(x):
    return (x.last_trip_date) - (x.signup_date) > timedelta(days=(5*366/12))  #leap year 
```

```python
ultimate['signup_date'] = pd.to_datetime(ultimate['signup_date'])
ultimate['last_trip_date'] = pd.to_datetime(ultimate['last_trip_date'])

ultimate['user_retension'] = ultimate.apply(user_retension, axis = 1)
```

```python
ultimate.user_retension.value_counts()
```

```python
ultimate
```

```python
# Fraction of observed users who were retained
user_retension_fraction = sum(ultimate.user_retension == True) / len(ultimate)
print('Fraction of observed users who were retained : ', user_retension_fraction)
```

22.38% of users are retained

```python
## Categorical column treatment
ultimate_cat = ultimate.copy()

ultimate_cat['city'].replace(["King's Landing", 'Astapor', 'Winterfell'] , [0, 1, 2], inplace=True)

ultimate_cat['phone'].replace(['iPhone', 'Android'] , [0, 1], inplace=True)

#datetime to ordinal
ultimate_cat['signup_date']=ultimate_cat['signup_date'].map(datetime.toordinal)

ultimate_cat['last_trip_date']=ultimate_cat['last_trip_date'].map(datetime.toordinal)
```

```python
from sklearn.model_selection import train_test_split

# data split 
X = ultimate_cat.iloc[:,:12]
y = ultimate_cat.iloc[:,12]

#test-train split in the ratio of 4:1
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 101)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import recall_score
from sklearn.metrics import confusion_matrix, accuracy_score, make_scorer

logmodel = LogisticRegression() 
logmodel.fit(X_train,y_train)

logpred_train = logmodel.predict(X_train)
logpred_test = logmodel.predict(X_test)
recall_train_logmod = recall_score(y_train, logpred_train)
recall_test_logmod = recall_score(y_test, logpred_test)


print("Training Accuracy score : ",round(accuracy_score(y_train, logpred_train),2)*100)
print("Confusion matrix : \n", confusion_matrix(y_train, logpred_train))
print("Recall - train : ", recall_train_logmod)


print("\nTesting Accuracy : ",round(accuracy_score(y_test, logpred_test),2)*100)
print("Confusion matrix : \n", confusion_matrix(y_test, logpred_test))
print("Recall - test : ", recall_test_logmod)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import model_selection

# Iterate to compute the optimal 'k'
X_trainK, X_testK, y_trainK, y_testK = train_test_split(X, y, test_size = 0.2, random_state = 101)

#Neighbors
neighbors = np.arange(0,25)

#Create empty list that will hold cv scores
cv_scores = []

#Perform 10-fold cross validation on training set for odd values of k:
for k in neighbors:
    k_value = k+1
    knn = KNeighborsClassifier(n_neighbors = k_value, weights='uniform', p=2, metric='euclidean')
    kfold = model_selection.KFold(n_splits=10)
    scores = model_selection.cross_val_score(knn, X_trainK, y_trainK, cv=kfold, scoring='accuracy')
    cv_scores.append(scores.mean()*100)
    print("k=%d %0.2f (+/- %0.2f)" % (k_value, scores.mean()*100, scores.std()*100))

optimal_k = neighbors[cv_scores.index(max(cv_scores)) + 1]
print ("The optimal number of neighbors is %d with %0.1f%%" % (optimal_k, cv_scores[optimal_k]))

plt.plot(neighbors, cv_scores)
plt.xlabel('Number of Neighbors K')
plt.ylabel('Train Accuracy')
plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors=20)
knn.fit(X_train, y_train)

knnpred_train = knn.predict(X_train)
knnpred_test = knn.predict(X_test)
recall_train_knn = recall_score(y_train, knnpred_train)
recall_test_knn = recall_score(y_test, knnpred_test)

print("Training Accuracy score : ",round(accuracy_score(y_train, knnpred_train),2)*100)
print("Confusion matrix : \n", confusion_matrix(y_train, knnpred_train))
print("Recall - train : ", recall_train_knn)

print("\nTesting Accuracy score : ",round(accuracy_score(y_test, knnpred_test),2)*100)
print("Confusion matrix : \n", confusion_matrix(y_test, knnpred_test))
print("Recall - test : ", recall_test_knn)
```
